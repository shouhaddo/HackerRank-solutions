```python
lst = []
scores = []
if __name__ == '__main__':
    for _ in range(int(input())):
        name = input()
        score = float(input())
        lst.append([name,score])
        scores.append(score)
    
    num = min(scores)
    while (min(scores) == num ) :
        scores.remove(min(scores))
    
    num = min(scores)
    lst.sort()

    for val1,val2 in lst :
        if val2 == num :
            print(val1)
```
